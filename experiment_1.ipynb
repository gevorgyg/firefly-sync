```python
import networkx as nx
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import random
import numpy as np
matplotlib.use('TkAgg')
# Set up the graph
G = nx.erdos_renyi_graph(20, 0.15)  # Random graph with 20 nodes
pos = nx.spring_layout(G)  # Positions for all nodes

# Create the figure and axis
fig, ax = plt.subplots(figsize=(10, 8))
plt.title('Network Graph with Changing Node Colors')

# Initial node colors (all blue to start)
node_colors = ['blue'] * len(G)

# Draw the initial graph
nodes = nx.draw_networkx_nodes(G, pos, node_color=node_colors, ax=ax)
edges = nx.draw_networkx_edges(G, pos, ax=ax)
labels = nx.draw_networkx_labels(G, pos, ax=ax)

def update(frame):
    global node_colors

    # Every 5 frames, update the colors
    if frame % 2 == 0:
        node_colors = [random.choice(['red', 'blue']) for _ in range(len(G))]

    # Update node colors
    nodes.set_array(np.array([1 if color == 'red' else 0 for color in node_colors]))
    nodes.set_cmap(plt.cm.bwr)  # Blue to red colormap

    # Optional: Update title with frame count
    ax.set_title(f'Network Graph - Frame {frame}')

    return nodes, edges, labels

# Create animation
ani = FuncAnimation(
    fig,
    update,
    frames=100,  # 100 frames total
    interval=200,  # 200ms between frames
    blit=False
)

plt.tight_layout()
plt.show()
```
